```python
import string

def remove_punc(str1):
    str2 = ""
    for char in str1:
        if char not in string.punctuation: 
            str2 += char
    return str2

def word_split(str2):
    words = str2.split()
    return words

def word_dict(L1):
        keys=[]
        values=[]
        for k in L1:
            if k not in keys:
                keys.append(k)
        for v in keys:
            count=L1.count(v)
            values.append(count)
        
        d1={k:v for k,v in zip(keys,values)}
    
        return d1

def word_freq_table(d1):
    import pandas as pd
    df=pd.DataFrame({'word':list(d1.keys()),'no of occurences':list(d1.values())})  
    return df

if __name__ == "__main__":
    str1='When indefinite articles are used before a noun (such as a ball), they do not refer to a specific noun (like the soccer ball) but a noun in general (any ball—maybe a tennis ball or a football or even a baseball) So if you tell your friend to grab a ball, she might bring back a tennis ball or a football or even a baseball because you were not specific about which ball you wanted.'
    str2 = remove_punc(str1)
    L1 = word_split(str2)
    d1= word_dict(L1)
    word_frequency_df = word_freq_table(d1)
    print(word_frequency_df)
```
