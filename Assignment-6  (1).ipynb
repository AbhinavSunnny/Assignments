```python
d='New data from the World Health Organization (WHO) and UN Children’s Fund UNICEF, reveal signs of immunisation services rebounding in some countries, but coverage still falls short of pre-pandemic levels, particularly in low-income countries, putting children at grave risk of disease outbreaks.'
```

```python
d1="".join([i for i in d if i not in [string.punctuation]])  # Taking Puctuations out
d1
```

```python
d2 = d1.split()  # Splitting the data
```

```python
d2   #The splitted data in the list
```

```python
keys=[]          # creating keys for dictionary
for k in d2:
    if k not in keys:
        keys.append(k)
keys
    
    
```

```python
values=[]       # creating values for dictionary
for m in keys:
    values.append(d2.count(m))
    
values
```

```python
dict1 = { k:v for (k,v) in zip(keys, values)}    # creating the dictionary using zip
dict1
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'word':list(dict1.keys()),'no of occurences':dict1.values()})  # creating the data frame
df
```
